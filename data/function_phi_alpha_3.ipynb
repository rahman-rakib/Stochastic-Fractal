```python
import numpy as np
import pandas as pd
import glob
import json
import os
```

```python
dir_data = "./phi_list/"
files = glob.glob(dir_data+"phi_*.csv")
```

```python
data_dict = dict()

data_dict['xi'] = np.linspace(0,2,10001)

for file in files:
    #print(file)
    a = file.split('_')[-2]
    #print(a)
    phi_list_a = np.loadtxt(file)
    col_name = 'p={}'.format(int(a)/100)
    #print(col_name)
    data_dict[col_name] = phi_list_a
```

```python
data_dict
```

```python
df_data = pd.DataFrame(data_dict)
```

```python
df_data
```

```python
def phi_value_alpha_3(xi,p):
    
    temp_p = float('{:.2f}'.format(p))
    del_p = abs(temp_p-p)
    
    if del_p <= 1e-9:
        col_name = 'p={:.2f}'.format(p)
        #print(col_name)
        phi_list = df_data[col_name].values
    
    else:
        col_name0 = 'p={:.2f}'.format(p)
        #print(col_name0)
        phi_list0 = df_data[col_name0].values
    
        col_name1 = 'p={:.2f}'.format(temp_p+0.01)
        #print(col_name1)
        phi_list1 = df_data[col_name1].values
        
        del_phi_list =  (phi_list1 - phi_list0)*del_p/0.01
        phi_list = phi_list0 + del_phi_list
        
    #print(del_phi_list)
    
    xi_list = df_data['xi'].values
    if xi > 2:
        phi_value = 0
    elif xi in xi_list:
        index = list(xi_list).index(xi)
        phi_value = phi_list[index] 
    else:
        index0 = int(xi//0.0002)
        index1 = int(index0+1)
        del_xi = xi%0.0002
        phi_0 = phi_list[index0]
        phi_1 = phi_list[index1]
        del_phi = (phi_1-phi_0)*del_xi/0.0002
        phi_value = phi_0 + del_phi

    return phi_value

def phi_list_alpha_3(xi_list,p):
    phi_list = []
    for xi in xi_list:
        phi_value = phi_value_alpha_3(xi,p)
        phi_list.append(phi_value) 

    return np.array(phi_list) 
```

```python
xi_list = np.linspace(0,2,700)
p = 0.999
density = phi_list_alpha_3(xi_list,p)
```

```python
from matplotlib import pyplot as plt
plt.plot(xi_list,density)
```
